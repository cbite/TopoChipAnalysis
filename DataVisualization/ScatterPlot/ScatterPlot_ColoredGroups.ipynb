<h3> Jupyter notebook to visualize biological readouts in scatter plot </h3>

<h5> Set the working directory and load libraries </h5>

```python
import os
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

<h5> Specify the file containing the biological readouts </h5> 
<p> The file should contain at least three columns: feature idx | readout 1 | readout 2. Of course, a file contain more columns if there are more biological readouts measured.</p>

```python
input_file_readouts=pd.read_csv('DataFrame_aSMA_EdU_Macrophage_PatternArea.csv',sep=',')
```

```python
input_file_readouts.columns
```

<h5> Select the columns to visualize in the scatterplot </h5>

```python
column_one='Nuclei_Count(aSMA)' 
column_two='Mean_Macrophage_count'
```

Pick a column to color and set the lower and upper boundary to create 3 groups

```python
column_to_color_points='Mean_Macrophage_count'
```

```python
lower_boundary=22.5
upper_boundary=27.5
```

```python
color_points=[]
for x in input_file_readouts[column_to_color_points]:
    if x <lower_boundary:
        color_points.append('Group 1')
    if (x < 27.5) and (x>=22.5):
        color_points.append('Group 2')
    if x >=27.5:
        color_points.append('Group 3')
```

<h5> Create the scatter plot </h5>

```python
plt.figure(figsize=(8.0, 5.0),facecolor=(1, 1, 1))
scatter_plot=sns.scatterplot(x=column_one,y=column_two,data=input_file_readouts,hue=color_points,hue_order = ['Group 3', 'Group 2','Group 1'], palette=['red','green','darkblue'])
```

Save the figure to a svg and 300-dpi file

```python
scatter_plot_figure=scatter_plot.get_figure()
scatter_plot_figure.savefig('Scatterplot.png', dpi=300)
```
