Load the first input file

```python
input_file_readouts=pd.read_csv('EdUScreen_EdURatio_sorted.txt',sep='\t')
```

<h5> Select the columns to visualize in the scatterplot </h5>

```python
column_one='Count_AllCells'
column_two='Count_EdUPositive'
```

<h5> Create the scatter plot </h5>

```python
sns.scatterplot(x=column_one,y=column_two,data=input_file_readouts)
```

<h4> Select the feature idx based on readout values </h4>

You can select the feature idx of interest based on the two biological readouts. First specify the value for column 1 (defined above) and column 2.

```python
column_one_condition='<'
column_two_condition='<'
```

```python
column_one_value=10
column_two_value=5
```

```python
subselection_data=input_file_readouts.loc[input_file_readouts[column_one]<column_one_value,:]
subselection_data=subselection_data.loc[subselection_data[column_two]<column_two_value,:]
```

<h5> Create a new scatter plot based on the conditions </h5>

```python
sns.scatterplot(x=column_one,y=column_two,data=subselection_data)
```

Rank the data in descending order based on the selected column

```python
column_to_sort='RatioEdUPositive'
```

```python
subselected_data_ranked=subselection_data.sort_values(column_to_sort,ascending=False)
```
